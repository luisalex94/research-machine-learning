### The np.arange() function

The np.arange() produces a sequence of numbes with equal increments.
* When only one input is passed, the sequence of numbers is from zero to the input number, with increments of one.
* When two inputs are passed, the sequence of numbers is from the first input to the second input, with increments of one.
* When three inputs are passed, the sequence of numbers is from the first input to the second input, with increments of the third input.

```python
x = np.arange(15)

# Display the type and content of x
print(f'Type of x:\n{type(x)}\n')
print(f'Content of x:\n{x}\n')

x = np.arange(5, 15)

# Display the type and content of x
print(f'Type of x:\n{type(x)}\n')
print(f'Content of x:\n{x}\n')

x = np.arange(-7.1, 7)

# Display the type and content of x
print(f'Type of x:\n{type(x)}\n')
print(f'Content of x:\n{x}\n')

x = np.arange(-7.1, 7, 0.5)

# Display the type and content of x
print(f'Type of x:\n{type(x)}\n')
print(f'Content of x:\n{x}\n')
```

### Example - Using a placeholder to accomodate analytics

### Pandas pivot and melt functions

In a nutshell, .pivot() and .melt() help you to switch between two forms of two dimensional data structures: wide form and long form.

The .melt() function can easily reshape a dataset from the wide form to the long form.

The .pivot() function can do the opposite.

##### The .melt() function

The .melt() function requires four inputs:

* id_vars: This input takes the indentifying columns.
* value_vars: This input takes the columns that hold the values.
* var_name: This input takes the name you would like to give to the identifying column that will be added to the long format.
* value_name: This input takes the name you would like to give to the new value column that will be added to the long format.

```python
wide_df = pd.read_csv("../data/wide/wide.csv")
wide_df
```

```python
wide_df.melt(
    id_vars="ReadingDateTime",
    value_vars=["NO", "NO2", "NOX", "PM10", "PM2.5"],
    var_name="Species",
    value_name="Value"
)
```

##### The .pivot() function

The .pivot() function reshapes a DataFrame from the long form to the wide form. This function requires three inputs:
* index: This input takes what will be the index of the wide form.
* columns: This input takes the columns of the long form that will be expanded to create the columns for the wide form.
* values: This input takes the column in which the long form keeps the values.

```python
long_df = pd.read_csv("../data/long/long.csv")
long_df
```

```python
long_df.pivot(
    index="ReadingDateTime",
    columns="Species",
    values="Value"
)
```
